```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import tqdm
from collections import defaultdict
import json
from gensim.models import word2vec
import multiprocessing
cpu = multiprocessing.cpu_count()
```

```python
with open("/content/drive/MyDrive/datas/h_and_m/custvec/result.json", "r") as f :
  result = json.load(f)
```

```python
with open("/content/drive/MyDrive/datas/h_and_m/custvec/custdict.json", "r") as f :
  custd = json.load(f)
```

```python
df = pd.read_csv("/content/drive/MyDrive/datas/h_and_m/sample_submission.csv")
```

```python
df.head()
```

## submissionを構成する

```python
pred = []

num = 0
for cust_id in tqdm.tqdm(df['customer_id']) :
  r = result[cust_id]
  l = []
  for id, _ in r :
    if id in custd :
      l += custd[id]
  l.sort(reverse=True)

  sel = set()
  rest = []
  for day, item in l :
    month = day[5:7]
    if month == '09' or month == '10' :
      sel.add(item)
    else :
      rest.append(item) 
    if len(sel) == 12 : break


  for item in rest :
    if len(sel) == 12 : break
    sel.add(item)

  pred.append(" ".join(sel))
  # num+=1
  # if num == 10 :  break
```

```python
df['prediction'] = pred
df.head(5)
```

```python
df.to_csv("/content/drive/MyDrive/datas/submission.csv",columns=['customer_id', 'prediction'], index=False)
```
